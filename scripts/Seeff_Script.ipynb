```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
c = 0
Dataset = []
for page in range(1, 14, 1): 
    url = f'https://www.seeff.co.bw/results/residential/for-sale/?p={page}'
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser')

    for Listing in bs.find_all('div', id =lambda x: x and x.startswith("id_property_residential")): 
        c += 1
        Agency = "Seeff"
        Title_Text = Listing.find('h4', class_ = 'property-marketing-heading clearfix sub-headings')
        if Title_Text:
            # Split the text at the first occurrence of "in"
            parts = Title_Text.get_text(strip=True).split(" in ", 1)

            # Assign the split parts to variables
            title = parts[0]
            location = parts[1]
        else:
            title = None
            location = None
        
        bedrooms_count = Listing.find('div', class_ = 'property-list-beds pull-left fifth')
        if bedrooms_count:
            bedrooms = bedrooms_count.get_text(strip=True)
        else:
            bedrooms = None 
        bathrooms_count = Listing.find('div', class_ = 'property-list-baths pull-left fifth')
        if bathrooms_count:
            bathrooms = bathrooms_count.get_text(strip=True)
        else:
            bathrooms = None 
        parking_count = Listing.find('div', class_ = 'property-list-garages pull-left fifth')
        if parking_count:
            parking_spaces = parking_count.get_text(strip=True)
        else:
            parking_spaces = None

        P_Size_count = Listing.find('div', class_ = 'value')
        if P_Size_count:
            property_size = P_Size_count.get_text(strip=True)
        else:
            property_size = None 

        Descrip = Listing.find('p', class_ = 'margin-top15 margin-bottom0 clearfix results-content')
        if Descrip:
            description = Descrip.get_text(strip=True)
        else:
            description = None

        Link_Div = Listing.find('div', class_ = 'property-list-details half pull-left')
        link_loc = Link_Div.find('a')
        if link_loc and 'href' in link_loc.attrs:
            link = f"https://www.seeff.co.bw{link_loc['href']}"
        else:
            link = None

        Price_Loc = Listing.find_next('h3', class_ = 'property-price-heading headings')
        if Price_Loc:
            price = Price_Loc.get_text(strip=True)
        else:
            price = None

        
        Data = {
            'Title': title,
            'Location': location,
            'Remax_Description':description,
            'P24_Link':link,
            'Bedrooms': bedrooms,
            'Bathrooms': bathrooms,
            'Parking Spaces': parking_spaces,
            'Erf Size': property_size,
            'Agency':Agency,
            'Price': price
        }
        Dataset.append(Data)

Dataset_Part1 = pd.DataFrame(Dataset)
print(len(Dataset_Part1))
```

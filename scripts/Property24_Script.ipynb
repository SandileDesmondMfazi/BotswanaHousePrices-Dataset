```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
def count_digits(n):
    return len(str(n))

# bs.find_all(class_= re.compile('UnfussyTexan'))
def has_six_characters(css_class):
    n = 114760940
    length = count_digits(n)
    return css_class is not None and len(css_class) == length
```

```python
c = 0
Dataset = []
for page in range(1, 24, 1): 
    url = f'https://www.property24.co.bw/property-for-sale-in-gaborone-c2250?Page={page}'
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser') 

    for Listing in bs.find_all(attrs={"data-listing-number": has_six_characters}): 
        c += 1
        Agency = "Property24"
        title = Listing.find('span', class_='p24_propertyTitle').get_text(strip=True)
        # FeaturesDetails = Listing.find('span', class_='p24_featureDetails')
        bedrooms_count = Listing.find(attrs={'title': 'Bedrooms'})
        if bedrooms_count:
            bedrooms = bedrooms_count.get_text(strip=True)
        else:
            bedrooms = None 
        bathrooms_count = Listing.find(attrs={'title': 'Bathrooms'})
        if bathrooms_count:
            bathrooms = bathrooms_count.get_text(strip=True)
        else:
            bathrooms = None 
        parking_count = Listing.find(attrs={'title': 'Bathrooms'})
        if parking_count:
            parking_spaces = parking_count.get_text(strip=True)
        else:
            parking_spaces = None

        P_Size_count = Listing.find(attrs={'title': 'Erf Size'})
        if P_Size_count:
            property_size = P_Size_count.get_text(strip=True)
        else:
            property_size = None 

        location = Listing.find('span',{'class':'p24_location'}).get_text(strip=True)
        descrip = Listing.find('span',{'class':'p24_excerpt'}).get_text(strip=True)
        link = f"https://www.property24.co.bw{Listing.find('a')['href']}"
        price = Listing.find('span',{'class':'p24_price'}).get_text(strip=True) 

        Data = {
            'Title': title,
            'Location': location,
            'P24_Description':descrip,
            'P24_Link':link,
            'Bedrooms': bedrooms,
            'Bathrooms': bathrooms,
            'Parking Spaces': parking_spaces,
            'Erf Size': property_size,
            'Agency':Agency,
            'Price': price
        }
        Dataset.append(Data)

Dataset_Part1 = pd.DataFrame(Dataset)
print(len(Dataset_Part1))
```

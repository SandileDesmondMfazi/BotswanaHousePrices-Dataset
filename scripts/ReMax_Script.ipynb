```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
c = 0
Dataset = []
for page in range(1, 10, 1): 
    url = f'https://www.remax.co.za/property/for-sale/botswana/?page={page}'
    response = requests.get(url)
    html = response.text
    bs = BeautifulSoup(html, 'html.parser')

    for Listing in bs.find_all('div', class_='outer-container'): 
        c += 1
        Agency = "Remax"
        Title_Text = Listing.find('div', class_ = 'property-card-info__suburb')
        if Title_Text:
            # Split the text at the first occurrence of "in"
            parts = Title_Text.get_text(strip=True).split(" in ", 1)

            # Assign the split parts to variables
            title = parts[0]
            location = parts[1]
        else:
            title = None
            location = None
        
        bedrooms_count = Listing.find(attrs={'itemprop': 'numberOfRooms'})
        if bedrooms_count:
            bedrooms = bedrooms_count.get_text(strip=True)
        else:
            bedrooms = None 
        bathrooms_count = Listing.find(attrs={'itemprop': 'amenityFeature'})
        if bathrooms_count:
            bathrooms = bathrooms_count.get_text(strip=True)
        else:
            bathrooms = None 
        parking_count = Listing.find_next_sibling(attrs={'itemprop': 'amenityFeature'})
        if parking_count:
            parking_spaces = parking_count.get_text(strip=True)
        else:
            parking_spaces = None

        P_Size_count = Listing.find('span')
        if P_Size_count:
            property_size = P_Size_count.get_text(strip=True)
        else:
            property_size = None 

        
        Descrip = Listing.find('p', class_ = 'property-card-info__description')
        if Descrip:
            description = Descrip.get_text(strip=True)
        else:
            description = None

        link_loc = Listing.find('a')
        if link_loc and 'href' in link_loc.attrs:
            link = f"https://www.remax.co.za{link_loc['href']}"
        else:
            link = None

        Price_Loc = Listing.find(attrs={'itemprop': 'price'})
        if Price_Loc:
            price = Price_Loc.get_text(strip=True)
        else:
            price = None

        
        Data = {
            'Title': title,
            'Location': location,
            'Remax_Description':description,
            'P24_Link':link,
            'Bedrooms': bedrooms,
            'Bathrooms': bathrooms,
            'Parking Spaces': parking_spaces,
            'Erf Size': property_size,
            'Agency':Agency,
            'Price': price
        }
        Dataset.append(Data)

Dataset_Part1 = pd.DataFrame(Dataset)
print(len(Dataset_Part1))
```

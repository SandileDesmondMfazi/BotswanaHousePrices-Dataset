```python
import pandas as pd
from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
c = 0
Dataset =[]
main_url = 'https://www.seeff.co.bw'
for page in range(1,9,1):
  html = urlopen(f'https://www.seeff.co.bw/results/residential/for-sale/gaborone/all/house/?p={page}&advanced_search=1')
  bs = BeautifulSoup(html.read(), 'html.parser')
  for property_link in bs.find_all('a', title=re.compile("Bedroom House For Sale")):
    Url = property_link.get('href')
    data = {
        'Url': main_url + Url
    }
    Dataset.append(data)

Dataset_Links = pd.DataFrame(Dataset)
Filtered = Dataset_Links.iloc[::2]
print(len(Filtered))
```

```python
Filtered = Filtered['Url'].to_list()
Filtered
```

```python
Seeff_Dataset = []
c = 1
for url in Filtered:
  html = urlopen(url)
  soup = BeautifulSoup(html.read(), 'html.parser')
  Links = soup.find_all('div',{'class':'full inline-block margin-top15'})
  for link in Links:
    title = soup.find('h1',{'class':'page-heading pull-left'}).get_text(strip= True)
    try:
      Bedrooms = link.find('div',string=re.compile('Bedrooms')).next_sibling.get_text(strip= True)
    except:
      Bedrooms = 'N/A'
    #try blocks are used to manage the error output the the scraper doesn't find the elements.
    try:
      Bathrooms = link.find('div',string=re.compile('Bathrooms')).next_sibling.get_text(strip= True)
    except:
      Bathrooms = 'N/A'
    try:
      Kitchens = link.find('div',string=re.compile('Kitchens')).next_sibling.get_text(strip= True)
    except:
      Kitchens = 'N/A'
    try:
      Reception_Rooms = link.find('div',string=re.compile('Reception Rooms')).next_sibling.get_text(strip= True)
    except:
      Reception_Rooms = 'N/A'
    try:
      Studies = link.find('div',string=re.compile('Studies')).next_sibling.get_text(strip= True)
    except:
      Studies = 'N/A'
    try:
      CarportsParkings = link.find('div',string=re.compile('Carports / Parkings')).next_sibling.get_text(strip= True)
    except:
      CarportsParkings = 'N/A'
    try:
      Security = link.find('div',string=re.compile('Security')).next_sibling.get_text(strip= True)
    except:
      Security = 'N/A'
    try:
      Flatlet = link.find('div',string=re.compile('Flatlet')).next_sibling.get_text(strip= True)
    except:
      Flatlet = 'N/A'
    try:
      DomesticAccommodation = link.find('div',string=re.compile('Domestic Accommodation')).next_sibling.get_text(strip= True)
    except:
      DomesticAccommodation = 'N/A'
    try:
      Pool = link.find('div',string=re.compile('Pool')).next_sibling.get_text(strip= True)
    except:
      Pool = 'N/A'
    try:
      Size = link.find('div',string=re.compile('Land Size')).next_sibling.get_text(strip= True)
    except:
      Size = 'N/A'

    Price = soup.find('span',{'id':'heading-price'}).get_text(strip= True)

    Dataset = {
        'Title': title,
        'Bedrooms': Bedrooms,
        'Bathrooms': Bathrooms,
        'Kitchens': Kitchens,
        'Reception_Rooms': Reception_Rooms,
        'Studies': Studies,
        'CarportsParkings': CarportsParkings,
        'Security': Security,
        'Flatlet': Flatlet,
        'DomesticAccommodation': DomesticAccommodation,
        'Pool': Pool,
        'Size': Size,
        'Price': Price
    }
    Seeff_Dataset.append(Dataset)
    c += 1
print(f'Scraped {c} Houses')
Dataset_Part2 = pd.DataFrame(Seeff_Dataset)
Dataset_Part2
```

```python
Dataset_Part2.drop_duplicates(subset = ['Title'], keep = 'first', inplace = True)
Dataset_Part2
```

```python
Dataset_Part2.to_csv('Seeff.csv', encoding='utf-8', index = True)
```
